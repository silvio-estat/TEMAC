Realizando o parseamento com o docling

```python
import os
from pathlib import Path
import logging
from docling.document_converter import DocumentConverter

# Configuração de log para aparecer no Jupyter
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def processar_pastas_iccrts(caminho_raiz_iccrts):
    """
    Percorre as pastas especificadas do ICCRTS, converte arquivos .pdf e .docx
    para Markdown e salva em pastas com sufixo _MD.
    """
    
    # Lista exata das pastas solicitadas
    pastas_alvo = [
        "ICCRTS_21_ano_2016",
        "ICCRTS_22_ano_2017",
        "ICCRTS_23_ano_2018",
        "ICCRTS_24_ano_2019",
        "ICCRTS_25_ano_2020",
        "ICCRTS_26_ano_2021",
        "ICCRTS_27_ano_2022",
        "ICCRTS_28_ano_2023",
        "ICCRTS_29_ano_2024",
        "ICCRTS_30_ano_2025"
    ]

    path_base = Path(caminho_raiz_iccrts)


    if not path_base.exists():
        logging.error(f"A pasta raiz '{path_base}' não foi encontrada. Verifique o caminho.")
        return

    logging.info("Inicializando o conversor Docling...")
    converter = DocumentConverter()

    for nome_pasta in pastas_alvo:
        dir_entrada = path_base / nome_pasta
        dir_saida = path_base / f"{nome_pasta}_MD"

        if not dir_entrada.exists():
            logging.warning(f"Pasta não encontrada: {dir_entrada}. Pulando...")
            continue

        dir_saida.mkdir(parents=True, exist_ok=True)
        logging.info(f"Processando pasta: {nome_pasta} -> Destino: {dir_saida.name}")

        arquivos = [f for f in dir_entrada.iterdir() if f.is_file() and f.suffix.lower() in ['.pdf', '.docx']]

        for arquivo in arquivos:
            try:
                logging.info(f"  Convertendo: {arquivo.name}")
                resultado = converter.convert(arquivo)
                texto_markdown = resultado.document.export_to_markdown()

                arquivo_saida = dir_saida / (arquivo.stem + ".md")

                with open(arquivo_saida, "w", encoding="utf-8") as f:
                    f.write(texto_markdown)

            except Exception as e:
                logging.error(f"  Erro ao converter '{arquivo.name}': {e}")

# O bloco __name__ == "__main__" funciona perfeitamente no Jupyter
if __name__ == "__main__":
    
    caminho_pasta2='/content/drive/My Drive/Mestrado/Mestrado PPCA/Dissertaçao/Artigos_C2_IC2Institute/ICCRTS'

    processar_pastas_iccrts(caminho_pasta2)
```

```python
import logging
from pathlib import Path
from docling.document_converter import DocumentConverter
from docling.datamodel.base_models import InputFormat
from docling.document_converter import PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions

# 1. Defina o seu caminho da pasta explicitamente
caminho_pasta = '/content/drive/My Drive/Mestrado/Mestrado PPCA/Dissertaçao/papers_baixados_temac'

def criar_converter_docling(usar_ocr=False, extrair_tabelas=True):
    """
    Cria um DocumentConverter com parâmetros de pipeline para PDF.
    """
    try:
        pdf_opts = PdfPipelineOptions()
        if hasattr(pdf_opts, "do_ocr"):
            pdf_opts.do_ocr = usar_ocr
        if hasattr(pdf_opts, "do_table_structure"):
            pdf_opts.do_table_structure = extrair_tabelas

        return DocumentConverter(
            format_options={
                InputFormat.PDF: PdfFormatOption(pipeline_options=pdf_opts)
            }
        )
    except Exception as e:
        logging.warning(f"Não foi possível aplicar opções avançadas ({e}). Usando configuração padrão.")
        return DocumentConverter()


def converter_pdfs_para_markdown(
    caminho_principal,
    nome_pasta_saida="PDFs_convertidos_MD",
    usar_ocr=False,
    extrair_tabelas=True
):
    # 2. Cria a nova pasta DENTRO de "papers_baixados_temac"
    caminho_saida = Path(caminho_principal) / nome_pasta_saida
    caminho_saida.mkdir(parents=True, exist_ok=True) 
    
    print(f"Pasta de saída garantida em: {caminho_saida}\n")

    # 3. Busca todos os PDFs na pasta principal
    pdfs = list(Path(caminho_principal).glob("*.pdf"))

    if not pdfs:
        print("Nenhum PDF encontrado para conversão.")
        return

    converter = criar_converter_docling(usar_ocr=usar_ocr, extrair_tabelas=extrair_tabelas)

    total = len(pdfs)
    ok = 0
    falhas = 0

    print(f"Iniciando a conversão de {total} arquivo(s)...\n")

    for pdf_path in sorted(pdfs):
        try:
            resultado = converter.convert(pdf_path) 
            markdown = resultado.document.export_to_markdown()

            # 4. Salva o novo arquivo .md na nova pasta
            arquivo_saida = caminho_saida / f"{pdf_path.stem}.md"
            with open(arquivo_saida, "w", encoding="utf-8") as f:
                f.write(markdown) 

            ok += 1
            print(f"[OK] Convertido: {pdf_path.name}")
        except Exception as e:
            falhas += 1
            print(f"[ERRO] Falha ao converter {pdf_path.name}: {e}")

    print(f"\nResumo: {ok}/{total} convertidos com sucesso, {falhas} falha(s).")
    print(f"Verifique seus arquivos em: {caminho_saida}")


# 5. Executa a função passando o seu caminho_pasta
converter_pdfs_para_markdown(
    caminho_principal=caminho_pasta,
    nome_pasta_saida="PDFs_convertidos_MD", # Nome da pasta que será criada dentro
    usar_ocr=False,       
    extrair_tabelas=True
)
```
