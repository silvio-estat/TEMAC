# Listagem de documentos baixados na library do NATO C2COE 

```python
import os
import pandas as pd
from pathlib import Path
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
BASE_PATH = '/content/drive/My Drive/Mestrado/Mestrado PPCA/Dissertaçao/Artigos_C2_OTAN/biblioteca_site'
```

```python
# Lê apenas arquivos (ignora pastas) dentro de BASE_PATH
arquivos = [
    nome for nome in os.listdir(BASE_PATH)
    if os.path.isfile(os.path.join(BASE_PATH, nome))
]

# Cria o DataFrame solicitado
df = pd.DataFrame({"nome_arquivo": sorted(arquivos)})

# Visualização rápida
df.head()
```

Gravando esse arquivo no drive

```python
output_base = Path(BASE_PATH)

excel_path  = output_base / 'info_documentos_nato_c2coe.xlsx'

df.to_excel(excel_path, index=False, engine='openpyxl')
```
