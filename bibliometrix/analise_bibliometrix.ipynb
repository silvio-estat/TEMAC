# Importação dos arquivos para análise bibliométrica

## 1. Importando a biblioteca de interesse

```r
library(bibliometrix)
```

## 2. Convertendo as bases de dados para o formato do bibliometrix

```r
# 1. Converter a base original da Scopus
M_scopus <- convert2df("ajustes_BD/base_scopus_consolidada.csv", dbsource = "scopus", format = "csv")
save(M_scopus, file = "M_scopus.rdata")
```

```r
# 2. Converter a base da IEEE (usando o dbsource "scopus" graças à nossa adaptação!)
M_ieee <- convert2df("ajustes_BD/base_ieee_adaptada_scopus.csv", dbsource = "scopus", format = "csv")
save(M_ieee, file = "M_ieee.rdata")
```

```r
# 3. Converter a base da Web of Science (formato de texto plano)
M_wos <- convert2df("ajustes_BD/base_wos_consolidada.txt", dbsource = "wos", format = "plaintext")
save(M_wos, file = "M_wos.rdata")
```

## 3. Juntando as três bases de dados

```r
colunas_chave <- c("CR", "AB", "DE", "ID") # References, Abstract, Author Keywords, Index Keywords

for (col in colunas_chave) {
  if (!col %in% names(M_ieee)) M_ieee[[col]] <- NA
  if (!col %in% names(M_scopus)) M_scopus[[col]] <- NA
  if (!col %in% names(M_wos)) M_wos[[col]] <- NA
}
```

```r
# 4. Juntar as três bases e remover publicações duplicadas entre elas
M_final <- mergeDbSources(M_scopus, M_wos, remove.duplicated = TRUE)
```

## 4. Salvando a base de dados

```r
#salvando o M_final em um arquivo rdata
save(M_final, file = "M_final.rdata")
```
