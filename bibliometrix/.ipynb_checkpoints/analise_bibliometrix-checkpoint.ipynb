## 2. Convertendo as bases de dados para o formato do bibliometrix

```python
# 1. Converter a base original da Scopus
M_scopus <- convert2df("ajustes_BD/base_scopus_consolidada.csv", dbsource = "scopus", format = "csv")
save(M_scopus, file = "M_scopus.rdata")
```

```python
# 2. Converter a base da IEEE (usando o dbsource "scopus" graças à nossa adaptação!)
M_ieee <- convert2df("ajustes_BD/base_ieee_adaptada_scopus.csv", dbsource = "scopus", format = "csv")
save(M_ieee, file = "M_ieee.rdata")
```

```python
# 3. Converter a base da Web of Science (formato de texto plano)
M_wos <- convert2df("ajustes_BD/base_wos_consolidada.txt", dbsource = "wos", format = "plaintext")
save(M_wos, file = "M_wos.rdata")
```

```python
load("M_scopus.rdata")
```

```python
head(M_scopus)
```

```python
# Contar quantos artigos têm "ZHANG Y" como autor a partir de 2016
alvo <- "ZHANG Y"

tem_autor <- sapply(M_scopus$AU, function(x) {
    if (is.na(x) || x == "") return(FALSE)
    autores <- trimws(unlist(strsplit(toupper(x), ";")))
    alvo %in% autores
})

artigos_zhang_y_2016 <- M_scopus[tem_autor & !is.na(M_scopus$PY) & M_scopus$PY >= 2016, ]

cat("Número de artigos de", alvo, "a partir de 2016:", nrow(artigos_zhang_y_2016), "\n")
```

```python
# Exibir os títulos dos artigos de ZHANG Y a partir de 2016
cat("Títulos dos artigos de", alvo, "a partir de 2016:\n\n")
for (i in 1:nrow(artigos_zhang_y_2016)) {
    cat(i, ". ", artigos_zhang_y_2016$TI[i], " (", artigos_zhang_y_2016$PY[i], ")\n", sep = "")
}
```

```python
# Calcula as citações locais
citacoes_locais <- localCitations(M_scopus, sep = ";")

# Mostra o top 10 dos autores com mais citações locais
head(citacoes_locais$Authors, 10)
```

```python
# Defina o nome do autor em primeira colocação
autor_top <- "BALAC N" 

# Roda a análise de impacto para esse autor específico
analise_autor <- Hindex(M_scopus, field = "author", elements = autor_top)

# Acessa a lista de artigos do autor (ela já vem ordenada por citações)
artigos_do_autor <- analise_autor$CitationList[[1]]

# Mostra apenas o primeiro resultado (o artigo mais citado)
head(artigos_do_autor, 1)
```
